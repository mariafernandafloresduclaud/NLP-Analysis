```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('stopwords')
nltk.download('punkt')

# Combine all stop words from the three languages
stop_words = set(stopwords.words('english')).union(stopwords.words('spanish')).union(stopwords.words('french'))

# Create function
def remove_stopwords (text):
    tokenized = word_tokenize(str(text))
    without_stopwords = [word for word in tokenized if not word in stop_words]
    return without_stopwords
```

```python
content_df["title"] = content_df["title"].apply(remove_stopwords)
content_df["description"] = content_df["description"].apply(remove_stopwords)
content_df["text"] = content_df["text"].apply(remove_stopwords)
```

```python
from nltk.stem import WordNetLemmatizer

def lemma(text):
    lemmatizer = WordNetLemmatizer() # Instantiate lemmatizer
    lemmatized = [lemmatizer.lemmatize(word) for word in text] # Lemmatize
    lemmatized_string = " ".join(lemmatized)
    return lemmatized_string

content_df["text"] = content_df["text"].apply(lemma)
content_df["title"] = content_df["title"].apply(lemma)
content_df["description"] = content_df["description"].apply(lemma)
```

```python
X = np.asarray(content_df["text"])
y = content_df["kind"]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import set_config; set_config("diagram")

# Create Pipeline see on lecture of machine larning day 6
# $CHALLENGIFY_BEGIN
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),  #this transforms text to numbers
    ('nb', MultinomialNB()),      #this is the model naives b. model - good in text because it classifies it
])

# $CHALLENGIFY_END
# Set parameters to search
# $CHALLENGIFY_BEGIN
parameters = {
    'tfidf__ngram_range': ((1,1), (2,2)),
    'tfidf__min_df': (0.01,0.05),
    'tfidf__max_df': (0.8,0.9),
    'nb__alpha': (0.01,0.1,1,10)
}

# $CHALLENGIFY_END see on model tuning and workflow on machine learning lectures
# Perform grid search on pipeline to search for best parameters
# $CHALLENGIFY_BEGIN
grid_search = GridSearchCV(
    pipeline, parameters, n_jobs=-1,
    verbose=1, scoring = "accuracy",
    refit=True, cv=5
)
grid_search.fit(X,y)
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()
content_arrays = count_vectorizer.fit_transform(content_df["text"].values)
content_arrays.toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer_text = TfidfVectorizer(max_df = 0.75, max_features = 5000, ngram_range=(1,2))
vectorized_reviews = pd.DataFrame(vectorizer_text.fit_transform(content_df["text"]).toarray(),
                                 columns = vectorizer_text.get_feature_names_out())

print(f" vectorized_reviews.shape = {vectorized_reviews.shape}")
```

```python
from sklearn.decomposition import LatentDirichletAllocation

n_components = 3

lda = LatentDirichletAllocation(n_components = n_components)
lda.fit(vectorized_reviews)
```

```python
document_mixture = lda.transform(vectorized_reviews)
document_mixture.shape
```

```python
round(pd.DataFrame(document_mixture, columns = [f"Topic {i+1}" for i in range(n_components)]),2)
```

```python
df = np.argmax(document_mixture, axis = 1)
df 
```

```python
topic_mixture = pd.DataFrame(lda.components_, columns = vectorizer_text.get_feature_names_out())
topic_mixture.shape
```

```python
topic_mixture
```

```python
def topic_word(vectorizer, model, topic, topwords, with_weights = True):
    topwords_indexes = topic.argsort()[:-topwords - 1:-1]
    if with_weights == True:
        topwords = [(vectorizer_text.get_feature_names_out()[i], round(topic[i],2)) for i in topwords_indexes]
    if with_weights == False:
        topwords = [vectorizer_text.get_feature_names_out()[i] for i in topwords_indexes]
    return topwords
```

```python
def print_topics(vectorizer, model, topwords):
    for idx, topic in enumerate(model.components_):
        print("-"*20)
        print("Topic %d:" % (idx))
        print(topic_word(vectorizer_text, model, topic, topwords))
```

```python
topic_word_mixture_text = [topic_word(vectorizer_text, lda, topic, topwords = 10, with_weights = False)for topic in lda.components_]
topic_word_mixture_text

print_topics(vectorizer_text, lda, topwords = 10)
```
